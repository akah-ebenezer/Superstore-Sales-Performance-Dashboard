```python
#now let us go for date conversion

df['order_date']= pd.to_datetime(df['order_date'])
```

```python
df.head()
```

```python
#from the above table, datetime has been converted. but let us still confirm from the following code

df['order_date'].dtype
```

```python
#let us now create 2 new column

df['year']=df['order_date'].dt.year
df['month']=df['order_date'].dt.month
```

```python
df.head()
```

```python
#lets ensure the year column is converted to integer

df['year'] = df['year'].astype(int)
```

```python
df.head()
```

```python
#let us check for sales trend over time

sales_trend= df.groupby('year')['sales'].sum()
sales_trend
```

```python
#let us visualise the sales trend over time

sales_trend.plot(kind='line',marker='o')
plt.title('sales trend over time')
plt.xlabel('year')
plt.ylabel('total sales')
plt.xticks(sales_trend.index.astype(int))
plt.show()
```

```python
#sales decreased steadily over the years from 20,000 in 2014 to 2,500 in 2017
```

```python
#let us check for most profitable category

df.groupby('category')['profit'].sum().sort_values(ascending=False)
```

```python
#office supplies generates the highest profit, making it the most profitable product.
```

```python
#lets check for sub category

df.groupby('sub_category')['profit'].sum().sort_values()
```

```python
#sub category such as bookcase, machines, appliances and tables generated negative profit
```

```python
#let us check for discount inpact

df[['discount','profit']].corr()
```

```python
#lets represent this in plot

sns.scatterplot(x='discount',y='profit',data=df)
plt.title('discount vs profit')
plt.show
```

```python
#let us check for top 10 customers

top_customers = df.groupby( 'customer_name')['sales'].sum().sort_values(ascending=False).head(10)
top_customers
```

```python
top_customers.plot(kind='bar')
plt.title('top 10 customers by sale')
plt.show
```

```python
#lets check region performance

df.groupby('region')[['sales','profit']].sum()
```

```python
#the central region generated the highest sales but recorded negative profit or high discount. While the west region geneated more profit
#making it the most profitable region.
```
